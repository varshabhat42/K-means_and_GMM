# **Kmeans**

```python
import csv
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix
from mlxtend.plotting import plot_confusion_matrix


def color(row):
  if row == 1:
    return 'darkred'
  elif row == 2:
    return 'green'
  else:
    return 'navy'

def label(row):
  if row == 'Ear_left':
    return 1
  elif row == 'Ear_right':
    return 2
  else:
    return 0


valid_rows = []
with open('cluster.txt', newline='') as csvfile:
  spamreader = csv.reader(csvfile, delimiter=' ')
  for row in spamreader:
    if(re.search('^[0-9]', row[0])):
      valid_rows.append(row)
df = pd.DataFrame(valid_rows, columns= ['X', 'Y', 'Class'])
df['X'] = pd.to_numeric(df['X'])
df['Y'] = pd.to_numeric(df['Y'])

df['label'] = df.apply (lambda row: label(row[2]), axis=1)
df['color'] = df.apply (lambda row: color(row[3]), axis=1)

#K-means calculation
X = np.column_stack((df.X, df.Y))
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
centers = kmeans.cluster_centers_
y_kmeans = kmeans.predict(X)
K=np.column_stack((df.X,y_kmeans))

#Scatter plot for actual data
plt.figure(figsize=(8, 6), dpi=80)
plt.scatter(df.X, df.Y, color = df.color)
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
plt.title('Plot for actual data with given labels')
plt.show()

#Scatter plot after K-means implementation
plt.clf()
print("\n\n")
df['y_kmeans'] = y_kmeans
df['color_pred'] = df.apply (lambda row: color(row[5]), axis=1)
plt.figure(figsize=(8, 6), dpi=80)
plt.scatter(X[:, 0], X[:, 1], c=df.color_pred)
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
plt.title('Plot using K-means with centroids')
plt.show()


#Plot confusion matrix
plt.clf()
print("\n\n") 
fig, ax = plot_confusion_matrix(conf_mat=confusion_matrix(df.label,y_kmeans), figsize=(8, 6), cmap=plt.cm.Blues)
plt.xticks(range(3), ['Head','Ear_left','Ear_right'])
plt.yticks(range(3), ['Head','Ear_left','Ear_right'])
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=20)
plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder
from scipy.stats import multivariate_normal

def plot_scatter_and_conf_matrix(param_gmm_labels, param_num_of_iters):
    #Scatter plot
    plt.clf()
    print("\n\n\n\n") 
    plt.figure(figsize=(8, 6), dpi=80)
    plt.scatter(X[:, 0], X[:, 1], c = df.gmm_color_pred)
    plt.scatter(mu[:, 0], mu[:, 1], c='black', s=200, alpha=0.5);
    plt.title("No. of iterations: " + str(param_num_of_iters+1))
    plt.show()

    #Confusion matrix plot
    plt.clf()
    print("\n") 
    fig, ax = plot_confusion_matrix(conf_mat=confusion_matrix(df.label,param_gmm_labels), figsize=(8, 6), cmap=plt.cm.Blues)
    plt.xticks(range(3), ['Head','Ear_left','Ear_right'])
    plt.yticks(range(3), ['Head','Ear_left','Ear_right'])
    plt.xlabel('Predictions', fontsize=18)
    plt.ylabel('Actuals', fontsize=18)
    plt.title('Confusion Matrix', fontsize=20)
    plt.show()

def calc_sigma(param_mu, param_y_k):
  sigma = []
  for i in range(len(param_mu)):
    temp_sig = 0
    for k in range(len(X)):
      diff = (X[k] - param_mu[i]).reshape(2,1)
      numerator = param_y_k[k,i]*(diff@diff.T)
      temp_sig+=numerator
    temp_sig= temp_sig/sum(param_y_k[:,i])
    sigma.append(temp_sig)
  return sigma


enc = OneHotEncoder(handle_unknown='ignore')
y_k = enc.fit_transform(y_kmeans.reshape(-1, 1)).toarray()

num_of_iters = 30
objective_function = []
for z in range(num_of_iters):

  #M-step

  #Calculate mu
  mu = np.dot(y_k.T, X)/ sum(y_k).reshape(3,1)

  #Calculate sigma
  sigma = calc_sigma(mu, y_k)

  #Calculate weights 
  weights = sum(y_k)/len(y_k)

  #Calculate joint gaussian pdf
  gauss_pdf = []
  for i in range(len(mu)):
    temp_pdf = multivariate_normal.pdf(X, mean= mu[i], cov=sigma[i])
    gauss_pdf.append(temp_pdf)
  gauss_pdf = np.array(gauss_pdf).T

  #E-step
  e_step_denominator= weights[0]*gauss_pdf[:, 0]+weights[1]*gauss_pdf[:, 1]+weights[2]*gauss_pdf[:, 2]
  #print(e_step_denominator)
  y_k_new = []
  for i in range(len(weights)):
    y_k_new.append(weights[i]*gauss_pdf[:, i]/e_step_denominator)
  y_k = np.stack((y_k_new[0],y_k_new[1],y_k_new[2]), axis = 1)


  #Add predicted values to data frame
  gmm_labels = np.argmax(y_k, axis=1)
  df['gmm_labels'] = gmm_labels
  df['gmm_color_pred'] = df.apply (lambda row: color(row[7]), axis=1)

  #Plots for iteration 1-4
  if z<4:
    plot_scatter_and_conf_matrix(gmm_labels, z)
  #Plots for last iteration
  elif z==num_of_iters-1:
    plot_scatter_and_conf_matrix(gmm_labels, z)
```

As observed from the confusion matrices, K means has 79 data points which are classified incorrectly, whereas in GMM it is just 2. The accuracy of GMM is hence much better. We can also observe the same from the scatter plots, the scatter plot of GMM after 30 iterations closely resembles the original data scatter plot, whereas K means has a lot of misclassified data.

The GMM clustering is iterative, it returns the probabilities of the data points belonging to each class instead of a direct prediction based off of the euclidean distance between data points. 

Conclusion: GMM performs much better than K means clustering.
